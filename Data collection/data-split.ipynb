```python
import pandas as pd
```

```python
df = pd.read_csv('biotech_stock_data.csv')
df.head()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
import random
from functools import reduce
```

```python
indices = [i for i in range(48)]
random.shuffle(indices)
```

```python
index_sets = [
    (indices[:25]),
    (indices[25:] + indices[:2]),
    (indices[12:37]),
    (indices[37:] + indices[:14])
]
```

```python
for i, index_set in enumerate(index_sets):
    temp_df = df.loc[index_set].reset_index(drop=True)
    temp_df.to_csv(f'testing_splits/set_{i}.csv')
```
